```python
# LOAD IN THE DATA

# Fetch housing data
import os
import tarfile
import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"
```

```python
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if (not os.path.isfile(os.path.join(housing_path, "housing.csv"))):
        os.makedirs(housing_path)
        tgz_path = os.path.join(housing_path, "housing.tgz")
        urllib.request.urlretrieve(housing_url, tgz_path)
        housing_tgz = tarfile.open(tgz_path)
        housing_tgz.extractall(path=housing_path)
        housing_tgz.close()
```

```python
fetch_housing_data()
```

```python
# Load into a pandas dataframe
import pandas as pd
```

```python
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

```python
# GENERATE TEST DATASET
# Make income category attribute to stratify split test dataset
import numpy as np

housing["income_cat"] = pd.cut(housing["median_income"],
                              bins=[0.0, 1.5, 3.0, 4.5, 6.0, np.inf],
                              labels=[1, 2, 3, 4, 5])
housing["income_cat"].hist()
```

```python
# Stratify split test dataset
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_indices, test_indices in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_indices]
    strat_test_set = housing.loc[test_indices]

print(strat_test_set["income_cat"].value_counts() / len(strat_test_set))
print(housing["income_cat"].value_counts() / len(housing))
```

```python
# Drop income category attribute from train and test
strat_train_set = strat_train_set.drop(["income_cat"], axis="columns")
strat_test_set = strat_test_set.drop(["income_cat"], axis="columns")
strat_train_set.head()
```

```python
# Bedrooms per room seems like a good feature to include so we add it to our strat train set
# Note - we'll also have to do this for the test set for when we evaluate the model

strat_train_set["bedrooms_per_room"] = strat_train_set["total_bedrooms"]/strat_train_set["total_rooms"]
strat_test_set["bedrooms_per_room"] = strat_test_set["total_bedrooms"]/strat_test_set["total_rooms"]
```

```python
# PREPARE DATA FOR ML

# Get fresh copy of data
housing = strat_train_set.drop(["median_house_value"], axis="columns")
housing_labels = strat_train_set["median_house_value"].copy()
```

```python
# Streamline data preprocessing by making pipeline for numerical
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])
```

```python
# Compose pipelines into one for all - numerical and categorical
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

num_attribs = list(housing.drop(["ocean_proximity"], axis="columns"))
cat_attribs = ["ocean_proximity"]

preprocessing_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", OneHotEncoder(), cat_attribs),
])

# Perform final fit transform on data with pipeline
housing_prepared = preprocessing_pipeline.fit_transform(housing)
```

```python
pd.DataFrame(housing_prepared).head()
```

```python
# SELECT AND TRAIN MODEL

# Train and Evaluate on Training set (simple Lin reg)

from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(housing_prepared, housing_labels)
```

```python
# Quick checking model performance
some_data = housing.iloc[:5]
some_data_labels = housing_labels.iloc[:5]
some_data_prepared = preprocessing_pipeline.transform(some_data)
print("Predictions: ", lin_reg.predict(some_data_prepared))
print("Real: ", some_data_labels.to_numpy())
```

```python
# Find RMSE score
from sklearn.metrics import mean_squared_error

housing_predictions = lin_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
# This isn't that great - the model is probably underfitting so it's time to try a more complex model
```

```python
# Trying DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(housing_prepared, housing_labels)
```

```python
# Get DecisionTreeRegressor RMSE

housing_predictions = tree_reg.predict(housing_prepared)
lin_mse = mean_squared_error(housing_labels, housing_predictions)
lin_rmse = np.sqrt(lin_mse)
lin_rmse
```

```python
# Oh shit this is overfitting terribly
# We gonna try cross validation to do this better
```

```python
from sklearn.model_selection import cross_val_score

def eval_model(model):
    scores = cross_val_score(model, housing_prepared, housing_labels, scoring="neg_mean_squared_error", cv=5)
    scores = np.sqrt(-scores)
    print("Scores", scores)
    print("Mean", scores.mean())
    print("Std", scores.std())

print("Tree Reg: \n")
eval_model(tree_reg)
```

```python
print("Linear Reg: \n")
eval_model(lin_reg)
```

```python
# Ok so currently decision tree is overfitting so bad that linear regression performs better.
# Let's try random forest
```

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor()

print("Forest Reg: \n")
# This next line takes a while
eval_model(forest_reg)
```

```python
from sklearn.svm import SVR
svm_reg = SVR(kernel="rbf", C=1)

print("SVM Reg: \n")
eval_model(svm_reg)
# svm_reg.fit(housing_prepared, housing_labels)
```

```python
#  Perform GridSearchCV on RandomForest

from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [32, 64, 128], 'max_features': [8, 9, 10]},
    {'bootstrap': [False], 'n_estimators': [10, 30], 'max_features': [2, 4, 8, 9, 10]},
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)

grid_search.fit(housing_prepared, housing_labels)
```

```python
# Get scores for all estimators
# Get best estimator and scores
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
print("\n\nBest Estimator", grid_search.best_estimator_)
print("Best Estimator Score: ", min(np.sqrt(-cvres["mean_test_score"])))
```

```python
# Try RandomizedSearchCV for better control over how much compute to allocate
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

random_params = {
    'n_estimators': randint(10, 200),
    'max_features': randint(10, 15)
}

random_grid_search = RandomizedSearchCV(forest_reg, random_params, n_iter=5)
random_grid_search.fit(housing_prepared, housing_labels)
```

```python

# Repeat showing stuff for random search
# Get scores for all estimators
# Get best estimator and scores
random_cvres = random_grid_search.cv_results_
for mean_score, params in zip(random_cvres["mean_test_score"], random_cvres["params"]):
    print(np.sqrt(mean_score), params)
print("\n\nBest Estimator", random_grid_search.best_estimator_)
print("Best Estimator Score: ", min(np.sqrt(-cvres["mean_test_score"])))
```

```python
# Select and retrain best model
model = random_grid_search.best_estimator_
model.fit(housing_prepared, housing_labels)
```

```python
# Looking at feature importances
feature_importances = model.feature_importances_
attribs = list(housing)
cat_encoder = preprocessing_pipeline.named_transformers_["cat"]
cat_attribs = list(cat_encoder.categories_)

attribs = attribs
print(attribs, "\n\n")
sorted(zip(feature_importances, attribs), reverse=True)
```

```python
# EVALUATION ON TEST DATASET
X_test = strat_test_set.drop("median_house_value", axis="columns")
y_test = strat_test_set["median_house_value"].copy()

X_test_prepared = preprocessing_pipeline.transform(X_test)

test_predictions = model.predict(X_test_prepared)
test_mse = mean_squared_error(y_test, test_predictions)
test_rmse = np.sqrt(test_mse)
print("Test set RMSE: ", test_rmse)
```

```python
# Now compute 95% confidence interval for RMSE
from scipy import stats
confidence = 0.95
mse_errors = (test_predictions - y_test)**2
np.sqrt(stats.t.interval(confidence, len(mse_errors) - 1,
       loc = mse_errors.mean(),
       scale = stats.sem(mse_errors)))
```

```python
print("Average error: ", np.average(np.absolute(np.array(y_test) - np.array(model.predict(X_test_prepared)))))
```
